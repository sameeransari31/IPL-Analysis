# Lets make a function jisme agar batsman ka naam string format mei de toh woh kis team k against sabse zayada runs banaye woh return karega

```python
def most_no_of_runs_against_team(batsman):
    batsman_deliveries = deliveries[deliveries['batsman'] == batsman]
    runs_against_teams = batsman_deliveries.groupby('bowling_team').sum('batsman_runs')['batsman_runs']
    most_runs_team = runs_against_teams.sort_values(ascending = False).head(1)
    return most_runs_team
```

```python
most_no_of_runs_against_team('V Kohli')
```

# Find the most destructive death over batsman in the history of IPL
### On the basis of given condition we will decide the most destructive death over batsman
### 1. Max strike rate in death over, Strike rate = (number of runs/number of balls)*100
### 2. min batsman 200 balls in 16 - 20 overs

```python
mask = deliveries['over'] >= 16
deathover = deliveries[mask]
deathover
balls = deathover.groupby('batsman')['batsman_runs'].count().sort_values(ascending =  False)
x = balls>200
batsman_list = balls[x].index.tolist()
```

for calculating strike rate we have to find 2 things
1. runs scored by all these 43 batsmen
2. balls played by these 43 batsmen

```python
final = deathover[deathover['batsman'].isin(batsman_list)]
```

```python
final
```

```python
runs = final.groupby('batsman').sum('batsman_runs')['batsman_runs'].sort_values(ascending = False)
runs
```

```python
balls = final.groupby('batsman')['ball'].count().sort_values(ascending =  False)
balls
```

```python
sr = (runs/balls)*100
```

```python
sr.sort_values(ascending = False)
```

# From this we can conclude that AB de Villiers is the most destructive death over batsman in the history of IPL till 2017
